## 1. Implement positive tagging in movie review corpus.

```
import nltk
nltk.download('all')
```

```
from nltk.corpus import movie_reviews
nltk.download('movie_reviews')
```

```
from nltk.tokenize import PunktSentenceTokenizer
nltk.download('punkt')
```

```
movie_reviews.categories()
```

```
movie_reviews.fileids('pos')
```

```
movie_review_pos = movie_reviews.raw('pos/cv975_10981.txt')
movie_review_pos
```

```
sentence_tokenizer = PunktSentenceTokenizer(movie_review_pos)
movie_review_pos_tokenize = sentence_tokenizer.tokenize(movie_review_pos)
movie_review_pos_tokenize
```

```
def process():
  for i in movie_review_pos_tokenize:
    words = nltk.word_tokenize(i)
    tag = nltk.pos_tag(words)
    print(tag)

process()
```

## 2. Calculate number of words in a positive movie review excluding stopwords.

```
from nltk.corpus import stopwords
stopword = stopwords.words('english')
stopword
```

```
wordcount = 0
for i in movie_review_pos_tokenize:
  words = nltk.word_tokenize(i)
  if words not in stopword:
    wordcount = wordcount+1

wordcount
```
